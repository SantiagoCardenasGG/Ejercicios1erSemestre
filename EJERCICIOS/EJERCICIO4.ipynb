```python
# EJERCICIO 4

# Un cuerpo posee una velocidad inicial de 12 m/s y una aceleración de 2 m/s2

# PROBLEMA

# ¿Cuánto tiempo tardará en adquirir una velocidad de 144 Km/h?

# HECHO POR

# Santiago Cárdenas

# VERSION

# 1.0
```

```python
#Declaracion de datos
veli=int(input('Ingrese el valor de veli:'))
velf=int(input('Ingrese el valor de velf:'))
aceleracion=int(input('Ingrese el valor de la aceleracion:'))
```

```python
#Conversion de km/h a m/s
velofin=velf*5/18
```

```python
#formulas aritmeticas
tiempo=(velofin-veli)/(aceleracion)
```

```python
#Resultado
print('El tiempo que se tarde es:',tiempo)
```
