```python
import numpy as np
#DECLARACION DE VARIABLES
ladoA=0.0
ladoB=0.0
ladoC=0.0
```

```python
#LECTURA DE DATOS
ladoA=float(input('Ingrese el valor del lado A:'))
ladoB=float(input('Ingrese el valor del lado B:'))
ladoC=float(input('Ingrese el valor del lado C:'))
```

```python
#PROCESOS
#CALCULA EL SEMIPERIMETRO
semiperimetro=(ladoA+ladoB+ladoC)/2
#condiciones
if(semiperimetro>ladoA,ladoB,ladoC):
  area=np.sqrt(semiperimetro*(semiperimetro-ladoA)*(semiperimetro-ladoB)*(semiperimetro-ladoC))
  
```

```python
print('El area es:',area)
```
