EJERCICIO 3

Una locomotora necesita 10 s. para alcanzar su velocidad normal que es 25m/s.
Suponiendo que su movimiento es uniformemente acelerado 

PREGUNTA

¿Qué aceleración se le ha comunicado y qué espacio ha recorrido antes de alcanzar la velocidad regular?

Jhair Sebatian Torres Moreno

VERSION

1.0

```python
#Declaracion de los datos
velf=int(input('Ingrese el valor de velf:'))
veli=int(input('Ingrese el valor de veli:'))
tiempo=int(input('Ingrese el valor del tiempo:'))
```

```python
#Procesos aritmeticos
aceleracion=(velf-veli)/(tiempo)
distancia=(veli*tiempo)+((aceleracion*tiempo**2)/(2))
print(distancia)
```

```python
#Resultados
print('La aceleracion es a= ',aceleracion)
print('La distancia es x= ',distancia)
```
