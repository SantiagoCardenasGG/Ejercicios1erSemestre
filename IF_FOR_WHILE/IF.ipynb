```python
# Descripcion

# Leer de un estudiante su nombre y las notas de los 3 parciales se necesita calcular su nota definitiva sabiendo que es el promedio de los parciales, ademas imprimir un concepto de aprobo o reprobo

# Autor: Santiago CÃ¡rdenas

# Fecha:1-09-22

# version: 1.0
```

```python
#Declaracion de Variables
var_ent_cad_nombreEstudiante = ""
var_ent_float_notaPP = 0.0
var_ent_float_notaSP = 0.0
var_ent_float_notaTP = 0.0
var_proSal_float_notaDefpro = 0.0
var_proSal_float_notaDefpor = 0.0
var_proSal_cad_concepto = ''
```

```python
#Lectura de Datos
var_ent_cad_nombreEstudiante = input('Digite nombre del estudiante: ');
var_ent_float_notaPP = float(input('Ingrese la primera nota: '))
var_ent_float_notaSP = float(input('Ingrese la segundo nota: '))
var_ent_float_notaTP = float(input('Ingrese la tercera nota: '))
```

```python
#Procesos
#Calculo de nota definitiva por promedio
var_proSal_float_notaDefpro = (var_ent_float_notaPP + var_ent_float_notaSP + var_ent_float_notaTP)/3

#Calculo de nota definitiva por porcentaje
var_proSal_float_notaDefpor = ((var_ent_float_notaPP*0.25) + (var_ent_float_notaSP*0.25 ) + (var_ent_float_notaTP*0.5))

# Calculo del concepto si aprueba
if(var_proSal_float_notaDefpro>=3.0):
  var_proSal_cad_concepto = 'Aprobo - Te mueves al siguiente curso'
# Calculo del concepto si reprueba
if(var_proSal_float_notaDefpro<=3.0):
  var_proSal_cad_concepto = 'Reprobo - Sigue intentando'
```

```python
#Salidas
print('Nombre del estudiante : ',var_ent_cad_nombreEstudiante)
print('Definitiva por promedio : ',var_proSal_float_notaDefpro)
print('concepto',var_proSal_cad_concepto)
print('Definitiva por porcentaje : ',var_proSal_float_notaDefpor)
```
