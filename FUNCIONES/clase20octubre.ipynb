```python
#Lectura de archivos excel
import pandas as pd
import matplotlib.pyplot as plt

archivo_excel = pd.read_excel('deptoAndres.xlsx')
n_dep = archivo_excel['Departamento']
pob = archivo_excel['Poblacion']
area = archivo_excel['Area']

#Funcion
def area_col():
  total_area = archivo_excel['Area'].sum()
  return total_area
print('El area total de colombia es: ',area_col())  


def prom_pers_d():
  total_p = archivo_excel['Poblacion'].sum()
  prom_p = total_p/len(n_dep)
  return prom_p

print('El promedio de personas por departamento es',prom_pers_d())

def depar_dat(nombre_dep):
  nombre_dep=input('Ingrese el nombre del departamento')

def numero_mayor():
  d_mayor=0
  for x in range(len(n_dep)):
    index_d= archivo_excel.loc[x,'Poblacion']
    if index_d>d_mayor:
      d_mayor=index_d
  return d_mayor

print('El departamento con mayor poblacion es: ',numero_mayor())
```
