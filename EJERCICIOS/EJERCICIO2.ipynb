```python
# EJERCICIO 2

# Un fórmula 1 que parte del reposo alcanza una velocidad de 216 km/h en 10 s.

# PROBLEMA

# Calcula su aceleración.

# HECHO POR

# Santiago Cárdenas

# VERSION

# 1.0
```

```python
#Declaracion de datos
velf=int(input('Ingrese el valor de velf:'))
veli=int(input('Ingrese el valor de veli:'))
#veli=0 ya que parte del reposo
tiempo=int(input('Ingrese el valor del tiempo:'))
```

```python
#Conversion de km/h a m/s
velofin=velf*5/18
```

```python
#Procesos aritmeticos
aceleracion=(velofin-veli)/(tiempo)
```

```python
#Resultado
print('El resultado de a=',aceleracion)
```
